```python
import pandas as pd
import datetime as dt

dfServices2020_2022 = pd.read_csv('Services2020_2022.csv')
dfServices2022_2024 = pd.read_csv('Services2022_2024.csv')

dfServices = pd.concat([dfServices2020_2022, dfServices2022_2024])

dfSuccess = pd.read_csv('Successes.csv')

dfServices['Service Date'] = pd.to_datetime(dfServices['Service Date'])
dfSuccess['Exit Date'] = pd.to_datetime(dfSuccess['Exit Date'])
```

```python
Housing = [
    'Permanent housing (other than RRH) for formerly homeless persons',
    'Owned by client, no ongoing housing subsidy',
    'Rental by client, no ongoing housing subsidy',
    'Rental by client, VASH housing Subsidy',
    'Rental by client, other ongoing housing subsidy',
    'Rental by client in a public housing unit',
    'Rental by client with RRH or equivalent subsidy',
    'Long-term care facility or nursing home',
    'Residential project or halfway house with no homeless criteria',
    'Owned by client, with ongoing housing subsidy',
    'Rental by client, with HCV voucher (tenant or project based)',
    'Rental by client, with GPD TIP housing subsidy'
    ]

# After services with SALT, how many clients transitioned into housing though any orgs in 2022

dfServices1 = dfServices.copy()

dfServices1 = dfServices1[dfServices1['Created by Organization']=='SALT Outreach Inc.']

dfServicesInTimeframe = dfServices1[(dfServices1['Service Date'] >= dt.datetime(2021, 1, 1)) & (dfServices1['Service Date'] <= dt.datetime(2022, 12, 31))]

dfSuccessInTimeframe = dfSuccess[(dfSuccess['Exit Date'] >= dt.datetime(2022, 1, 1)) & (dfSuccess['Exit Date'] <= dt.datetime(2022, 12, 31))]

dfServices1 = dfServices1.sort_values('Service Date').drop_duplicates('Client ID', keep='first')

dfSuccess1 = dfSuccessInTimeframe[dfSuccessInTimeframe['Exit Destination'].isin(Housing)]
dfSuccess1 = dfSuccess1.sort_values('Exit Date').drop_duplicates('Client ID', keep='last')

dfIndirectSuccess = pd.merge(dfServices1, dfSuccess1, on='Client ID', how='inner')

dfIndirectSuccess['Exit Date'] = pd.to_datetime(dfIndirectSuccess['Exit Date'])
dfIndirectSuccess['Service Date'] = pd.to_datetime(dfIndirectSuccess['Service Date'])

dfIndirectSuccess = dfIndirectSuccess[dfIndirectSuccess['Exit Date'] > dfIndirectSuccess['Service Date']]
dfIndirectSuccess = dfIndirectSuccess[dfIndirectSuccess['Exit Date'] - dfIndirectSuccess['Service Date'] <= pd.Timedelta(365, 'D')]

print('After services with SALT, how many clients transitioned into housing though any orgs in 2022')
print(dfIndirectSuccess['Client ID'].nunique(), 'clients transitioned into housing through any orgs in 2022')
print()

# Number of clients who were directly housed by our case management team in 2022

dfSuccessInTimeframe = dfSuccess[(dfSuccess['Exit Date'] >= dt.datetime(2022, 1, 1)) & (dfSuccess['Exit Date'] <= dt.datetime(2022, 12, 31))]

dfSuccess2 = dfSuccessInTimeframe[dfSuccessInTimeframe['Exit Destination'].isin(Housing)]

dfSuccess2 = dfSuccess2[dfSuccess2['Agency']=='SALT Outreach Inc.']

print('Number of clients who were directly housed by our case management team in 2022')
print(dfSuccess2['Client ID'].nunique(), 'clients were directly housed by our case management team in 2022')
print()

# Percent of people served in 2022 that transitioned to case management

dfServices3 = dfServices.copy()

dfServices3 = dfServices3[dfServices3['Created by Organization']=='SALT Outreach Inc.']

dfServicesInTimeframe = dfServices3[(dfServices3['Service Date'] >= dt.datetime(2021, 1, 1)) & (dfServices3['Service Date'] <= dt.datetime(2022, 12, 31))]

dfServicesInTimeframe = dfServicesInTimeframe.sort_values('Service Date').drop_duplicates('Client ID', keep='first')

dfCaseManagement = dfServices3[dfServices3['Service']=='Case Management']

dfCaseManagement = dfCaseManagement[(dfCaseManagement['Service Date'] >= dt.datetime(2022, 1, 1)) & (dfCaseManagement['Service Date'] <= dt.datetime(2022, 12, 31))]

dfCaseManagement = dfCaseManagement.rename(columns={'Service Date': 'Case Management Date'})

dfMerge = pd.merge(dfServicesInTimeframe, dfCaseManagement, on='Client ID', how='inner')

dfMerge['Case Management Date'] = pd.to_datetime(dfMerge['Case Management Date'])
dfMerge['Service Date'] = pd.to_datetime(dfMerge['Service Date'])

dfMerge = dfMerge[dfMerge['Case Management Date'] > dfMerge['Service Date']]

print('Percent of people served in 2022 that transitioned to case management')
print(round((dfMerge['Client ID'].nunique() / dfServicesInTimeframe['Client ID'].nunique())*100, 2), '% of people served in 2022 transitioned to case management')
print(round((dfMerge['Client ID'].nunique() / dfCaseManagement['Client ID'].nunique())*100, 2), '% of people served by case management in 2022 transitioned came from other services')
print()

# After services with SALT, how many clients transitioned into housing though any orgs in 2021

dfServices1 = dfServices.copy()

dfServices1 = dfServices1[dfServices1['Created by Organization']=='SALT Outreach Inc.']

dfServicesInTimeframe = dfServices1[(dfServices1['Service Date'] >= dt.datetime(2020, 1, 1)) & (dfServices1['Service Date'] <= dt.datetime(2021, 12, 31))]

dfSuccessInTimeframe = dfSuccess[(dfSuccess['Exit Date'] >= dt.datetime(2021, 1, 1)) & (dfSuccess['Exit Date'] <= dt.datetime(2021, 12, 31))]

dfServices1 = dfServices1.sort_values('Service Date').drop_duplicates('Client ID', keep='first')

dfSuccess1 = dfSuccessInTimeframe[dfSuccessInTimeframe['Exit Destination'].isin(Housing)]
dfSuccess1 = dfSuccess1.sort_values('Exit Date').drop_duplicates('Client ID', keep='last')

dfIndirectSuccess = pd.merge(dfServices1, dfSuccess1, on='Client ID', how='inner')

dfIndirectSuccess['Exit Date'] = pd.to_datetime(dfIndirectSuccess['Exit Date'])
dfIndirectSuccess['Service Date'] = pd.to_datetime(dfIndirectSuccess['Service Date'])

dfIndirectSuccess = dfIndirectSuccess[dfIndirectSuccess['Exit Date'] > dfIndirectSuccess['Service Date']]
dfIndirectSuccess = dfIndirectSuccess[dfIndirectSuccess['Exit Date'] - dfIndirectSuccess['Service Date'] <= pd.Timedelta(365, 'D')]

print('After services with SALT, how many clients transitioned into housing though any orgs in 2021')
print(dfIndirectSuccess['Client ID'].nunique(), 'clients transitioned into housing through any orgs in 2021')
print()

# Number of clients who were directly housed by our case management team in 2021

dfSuccessInTimeframe = dfSuccess[(dfSuccess['Exit Date'] >= dt.datetime(2021, 1, 1)) & (dfSuccess['Exit Date'] <= dt.datetime(2021, 12, 31))]

dfSuccess2 = dfSuccessInTimeframe[dfSuccessInTimeframe['Exit Destination'].isin(Housing)]

dfSuccess2 = dfSuccess2[dfSuccess2['Agency']=='SALT Outreach Inc.']

print('Number of clients who were directly housed by our case management team in 2021')
print(dfSuccess2['Client ID'].nunique(), 'clients were directly housed by our case management team in 2021')
print()

# Percent of people served in 2021 that transitioned to case management

dfServices4 = dfServices.copy()

dfServices4 = dfServices4[dfServices4['Created by Organization']=='SALT Outreach Inc.']

dfServicesInTimeframe = dfServices4[(dfServices4['Service Date'] >= dt.datetime(2020, 1, 1)) & (dfServices4['Service Date'] <= dt.datetime(2021, 12, 31))]

dfServicesInTimeframe = dfServicesInTimeframe.sort_values('Service Date').drop_duplicates('Client ID', keep='first')

dfCaseManagement = dfServices4[dfServices4['Service']=='Case Management']

dfCaseManagement = dfCaseManagement[(dfCaseManagement['Service Date'] >= dt.datetime(2021, 1, 1)) & (dfCaseManagement['Service Date'] <= dt.datetime(2021, 12, 31))]

dfCaseManagement = dfCaseManagement.rename(columns={'Service Date': 'Case Management Date'})

dfMerge = pd.merge(dfServicesInTimeframe, dfCaseManagement, on='Client ID', how='inner')

dfMerge['Case Management Date'] = pd.to_datetime(dfMerge['Case Management Date'])
dfMerge['Service Date'] = pd.to_datetime(dfMerge['Service Date'])

dfMerge = dfMerge[dfMerge['Case Management Date'] > dfMerge['Service Date']]

print('Percent of people served in 2021 that transitioned to case management')
print(round((dfMerge['Client ID'].nunique() / dfServicesInTimeframe['Client ID'].nunique())*100, 2), '% of people served in 2021 transitioned to case management')
print(round((dfMerge['Client ID'].nunique() / dfCaseManagement['Client ID'].nunique())*100, 2), '% of people served by case management in 2021 transitioned came from other services')
print()

# How many clients were housed through any orgs after services from SALT, Q1 2024

dfServices5 = dfServices.copy()

dfServices5 = dfServices5[dfServices5['Created by Organization']=='SALT Outreach Inc.']

dfServicesInTimeframe = dfServices5[(dfServices5['Service Date'] >= dt.datetime(2023, 1, 1)) & (dfServices5['Service Date'] < dt.datetime(2024, 3, 1))]

dfSuccessInTimeframe = dfSuccess[(dfSuccess['Exit Date'] >= dt.datetime(2024, 1, 1)) & (dfSuccess['Exit Date'] < dt.datetime(2024, 3, 1))]

dfServices5 = dfServices5.sort_values('Service Date').drop_duplicates('Client ID', keep='first')

dfSuccess5 = dfSuccessInTimeframe[dfSuccessInTimeframe['Exit Destination'].isin(Housing)]
dfSuccess5 = dfSuccess5.sort_values('Exit Date').drop_duplicates('Client ID', keep='last')

dfIndirectSuccess = pd.merge(dfServices5, dfSuccess5, on='Client ID', how='inner')

dfIndirectSuccess['Exit Date'] = pd.to_datetime(dfIndirectSuccess['Exit Date'])
dfIndirectSuccess['Service Date'] = pd.to_datetime(dfIndirectSuccess['Service Date'])

dfIndirectSuccess = dfIndirectSuccess[dfIndirectSuccess['Exit Date'] > dfIndirectSuccess['Service Date']]
dfIndirectSuccess = dfIndirectSuccess[dfIndirectSuccess['Exit Date'] - dfIndirectSuccess['Service Date'] <= pd.Timedelta(365, 'D')]

print('After services with SALT, how many clients transitioned into housing though any orgs in Q1 2024')
print(dfIndirectSuccess['Client ID'].nunique(), 'clients transitioned into housing through any orgs in Q1 2024')
print()
```
